```python
from netCDF4 import Dataset
import numpy as np
from pathlib import Path
import sys, getopt
import os
import argparse
from datetime import datetime
from datetime import timedelta
from datetime import date
```

```python
#Beit Dagan
lat = 34.81
lon = 32.00

interval = 0.05
latStart = -59.975
lonStart = -59.975
```

```python
#find nearest lat/lon
def calc_index_by_lat_lon(lat, lon):
    
    #find nearest lat/lon
    latOffset =  (lat+59.975)/interval
    lonOffset =  (lon+59.975)/interval
    
    return(round(latOffset), round(lonOffset))
```

```python
def get_file_date(fileName):
    #extract last part in the file name which is the date
    filename = os.path.basename(fileName)
    tempTuple = os.path.splitext(filename)
    filename = tempTuple[0]
    fileNameArr = np.array(filename.split("-"))
    filedate = fileNameArr[len(fileNameArr)-1]
    
    #the format is 222001010900Z YYYYMMDDHHMM
    year = int(filedate[:4])
    month = int(filedate[4:6])
    day = int(filedate[6:8])
    hour = int(filedate[8:10])
    min = int(filedate[10:12])
    
    return datetime(year, month, day, hour, min)
```

```python
def get_file_day_and_hour(fileName):
    #extract last part in the file name which is the date
    filename = os.path.basename(fileName)
    tempTuple = os.path.splitext(filename)
    filename = tempTuple[0]
    fileNameArr = np.array(filename.split("-"))
    filedate = fileNameArr[len(fileNameArr)-1]
    
    #the format is 222001010900Z YYYYMMDDHHMM
    year = int(filedate[:4])
    month = int(filedate[4:6])
    day = int(filedate[6:8])
    hour = int(filedate[8:10])
    min = int(filedate[10:12])
    
    return (day, hour)
```

```python
inputDir = '/Users/alonhrl/Documents/Alon/MA-MSC/development/eumetsat'

if(inputDir == None):
    inputDir = os. getcwd()
print(inputDir)

cLat, cLon = calc_index_by_lat_lon(lat, lon)

#get ssi for specified coordinates from each file 
files = Path(inputDir).glob('*.nc')
for file in files:
    fh = Dataset(file, mode='r')
    ssi = fh['ssi'][:]
    print("time: %s, latIndx: %d, lonIndx: %d,  ssi: %.2f" %(get_file_date(file), cLat, cLon, ssi[cLat,cLon]))
```

```python
# iterate recursively
import glob
inputDir = '/Users/alonhrl/Documents/Alon/MA-MSC/development/Eumetsat data'
#inputDir = '/Users/alonhrl/Documents/Alon/MA-MSC/development/eumetsat'
days_in_month = 31
num_of_radiation_hours = 24
#samples_start_hour = 8

#calculate indices for required lattitude & longitude
cLat, cLon = calc_index_by_lat_lon(lat, lon)

a_shape = (days_in_month+1, num_of_radiation_hours+1)  # we add one so the cell index will match day and hour
monthly_rad = np.zeros(a_shape, dtype=np.int16)

count = 0
for filename in glob.iglob(inputDir + '**/**', recursive=True):
    file_name, file_extension = os.path.splitext(filename)
    if file_extension == '.nc':
        count +=1
        # Push radiation value into the month array
        fh = Dataset(filename, mode='r')
        ssi = fh['ssi'][:]
        print("time: %s, latIndx: %d, lonIndx: %d,  ssi: %.2f" %(get_file_date(filename), cLat, cLon, ssi[cLat,cLon]))
        day, hour = get_file_day_and_hour(filename)
                
        #if value is 'nan' set to 0
        if type(ssi[cLat,cLon]) is not np.float64:
            monthly_rad[day][hour] = 0
        else:
            monthly_rad[day][hour] = ssi[cLat,cLon]

#remove first row (no day '0')
monthly_rad = monthly_rad[1:]

print (monthly_rad)        
print ("Processed %d files!" % (count))
```
